```python
from tensorflow.keras.applications import VGG19
from tensorflow.keras.layers import Input, Conv2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
from tensorflow.keras.losses import Huber
from keras.layers import Flatten, Dense, LeakyReLU
from tensorflow.keras.losses import mse

# Load the pre-trained ResNet50 model
base_model = VGG19(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Add bbox output layer
inputs = Input(shape=(224, 224, 3))
x = base_model(inputs)

# flatten the max-pooling output of ResNet50
flatten = Flatten()(x)
# construct a fully-connected layer header to output the predicted
# bounding box coordinates
bboxHead = Dense(128, activation="relu")(flatten)
bboxHead = LeakyReLU(alpha=0.1)(bboxHead)
bboxHead = Dense(64, activation="relu")(bboxHead)
bboxHead = LeakyReLU(alpha=0.1)(bboxHead)
bboxHead = Dense(32, activation="relu")(bboxHead)
bboxHead = LeakyReLU(alpha=0.1)(bboxHead)
bboxHead = Dense(4, activation="relu")(bboxHead)
bboxHead = LeakyReLU(alpha=0.1)(bboxHead)

# construct the model we will fine-tune for bounding box regression
model = Model(inputs=inputs, outputs=bboxHead)

# Freeze the weights of the pre-trained layers
for layer in base_model.layers:
    layer.trainable = True


# Compile the model
optimizer = Adam(learning_rate=1e-4)
model.compile(optimizer=optimizer, loss=mse)
```

```python
import tensorflow as tf
from tensorflow import keras

# Create a sample model
model = keras.Sequential([
    keras.layers.Input(shape=(784,)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

# Assume 'output_layer' is the last layer in the model
output_layer = model.layers[-1]

# Create a functional model to access intermediate layers
model_with_intermediates = keras.Model(inputs=model.input, outputs=output_layer.output)

# Start tracing back from the output layer
layers_backtracked = [output_layer]
model_with_intermediates.input

while layers_backtracked[-1] != model_with_intermediates.input:
    prev_layer_output = layers_backtracked[-1].input
    prev_layer = prev_layer_output.op.inputs[0].owner.layer
    layers_backtracked.append(prev_layer)

# # Reverse the list to get the tracing from input to output
# layers_backtracked.reverse()

# # Print the layers traced back
# for layer in layers_backtracked:
#     print(layer)
```
