```python
# MP4 to Images
import cv2

# Open the MP4 video file
video_path = 'df_result.mp4'
video = cv2.VideoCapture(video_path)

# Initialize frame count
frame_count = 0

# Read and save frames until the video ends
while video.isOpened():
    # Read the current frame
    ret, frame = video.read()

    # If reading was successful, save the frame as an image
    if ret:
        # Save the frame as an image file
        image_path = f'images/mrcnn_target/frame_{frame_count}.jpg'
        cv2.imwrite(image_path, frame)

        # Increment frame count
        frame_count += 1
    else:
        break

# Release the video object and close the file
video.release()
```

```python
# MP4 + MP3 -> new MP4
import re
from moviepy.editor import ImageSequenceClip, AudioFileClip

ROOT_DIR = os.path.abspath("../")

IMAGE_DIR = os.path.join(ROOT_DIR, 'Mask_RCNN', 'augmentation')
image_file_name = next(os.walk(IMAGE_DIR))[2]


def extract_number(string):
    matches = re.findall(r'\d+', string)
    if matches:
        return int(matches[-1])
    else:
        return 0

sorted_list = sorted(image_file_name, key=extract_number)

augmented_images = ['augmentation/'+file_name for file_name in sorted_list]


# Load the sequence of images as a video clip
image_sequence = ImageSequenceClip(augmented_images, fps=24)

# Load the audio clip
audio = AudioFileClip('When Asian kids get a B.mp3')

# Set the audio duration to match the video duration
audio = audio.set_duration(image_sequence.duration)

# Set the audio as the audio track of the video
video_with_audio = image_sequence.set_audio(audio)

# Write the final video with audio to a file
video_with_audio.write_videofile('output.mp4', codec='libx264')
```

```python
# add text to images
from PIL import Image, ImageDraw, ImageFont

for image in images:
    image_path = 'augmentation/%s' % image  # Path to the input image file
    output_path = 'edit_output/%s' % image  # Path to the output image file

    # Load the image
    image = Image.open(image_path)

    # Create a drawing object
    draw = ImageDraw.Draw(image)

    # Define the text to be added
    text1 = "Emotional Damage"
    text2 = "Asian Parents"

    # Define the font size and font file path
    font_size = 180
    font_file = 'arial.ttf'  # Path to the font file

    # Load the font
    font = ImageFont.truetype(font_file, font_size)

    # Determine the position to place the text (e.g., bottom-left corner)
    text_position1 = (50, image.height - font_size - 150)
    text_position2 = (180, image.height - font_size - 1250)

    # Specify the color of the text (RGB format)
    text_color = (0, 0, 0)

    # Add the text to the image
    draw.text(text_position1, text1, font=font, fill=text_color)
    draw.text(text_position2, text2, font=font, fill=text_color)

    # Save the modified image
    image.save(output_path)
```

```python
import moviepy.editor as mp
import os

def extract_images_with_subtitle(video_path, output_folder):
    # Load the video clip
    video_clip = mp.VideoFileClip(video_path)

    # Extract the frames as images
    frames = video_clip.iter_frames()
    frame_per_sec = len(list(video_clip.iter_frames()))/video_clip.duration
    print('frame per sec: %s' % frame_per_sec)

    # Create the output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # Iterate over the frames and save them as images
    for i, frame in enumerate(frames):
        image_path = os.path.join(output_folder, f"image_{i}.jpg")
        mp.ImageClip(frame).save_frame(image_path)

video_path = "output.mp4"
output_folder = "augmentation"
extract_images_with_subtitle(video_path, output_folder)
```
